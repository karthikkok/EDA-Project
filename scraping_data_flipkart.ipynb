```python
brand = []
model = []
processor = []
ram = []
memory = [] 
ratings = []
reviews = []
ram_version = []
display_inch = []
warranty = []
price = []

for i in range(1,80):
    
    url = f"https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
     
    headers = {"Accept-Language" : "en-US, en; q=0.9"}
    page = requests.get(url, headers=headers)
    
    soup = BeautifulSoup(page.text)

    for i in soup.find_all('div',class_='col col-7-12'):
        
        brand.append(''.join(re.findall(r'(^\w+)\s',i.text)))
        model.append(''.join(re.findall(r'^\w+\s(\w+\s?[\w.+\d.+]\w.+?[\s\d+]\w+)',i.text)))
        processor.append(''.join(re.findall(r'(\w+\s\w+\s\w+\s\w+\s\w+)\s\-',i.text)))
        ram.append(''.join(re.findall(r'\((\d+)\sGB',i.text)))
        memory.append(''.join(re.findall(r'\(\d+\sGB\/(\d+)',i.text)))
        ratings.append(''.join(re.findall(r'\/?(\w\.\d.+)Ratings',i.text)))
        reviews.append(''.join(re.findall(r'(\d+)\sReviews',i.text)))
        ram_version.append(''.join(re.findall(r'(DDR\w+)\sRAM',i.text)))
        display_inch.append(''.join(re.findall(r'\((\d+[\s\.]?\d)\s[Ii]nch',i.text)))
        warranty.append(''.join(re.findall(r'[\w+](\d+)\s[yY]?ear',i.text)))

    for i in soup.find_all('div',class_='Nx9bqj _4b5DiR'):
    
        price.append(''.join(re.sub(r'[₹,]','',i.text)))
```
